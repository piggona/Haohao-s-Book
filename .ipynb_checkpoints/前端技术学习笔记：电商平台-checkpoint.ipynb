# 架构设计

## 分层架构：纵向层级的划分
- 定义：把功能相似，抽象级别相近的实现进行分层隔离。
- 优势：松散耦合（易维护、易复用、易扩展）
- 常见分层方式：MVC, MVVM

## 模块化：横向模块的划分
- 定义：解决复杂问题，将系统部分分解为模块进行分块开发
- 意义：解耦，可以并行开发
- 模块化方案：AMD，CMD，CommonJS，ES6
### AMD:Jquery的模块化方案
将业务代码与模块化代码掺杂到一起
### CMD:CJs的模块化方案
将业务代码与模块化代码掺杂到一起
### *CommonJS：nodejs的模块化方案
语法不多，模块代码与业务分开。
### ES6：兼容性不高

## 模块设计

# Http代理
## 原理


**输入a.com的地址，但代理服务器进行劫持之后就将请求指向b.com**

## 开发时的作用:解决跨域问题

我们将前后端部署在不同的域名下，当我们请求前端文件时（a.com），我们直接指向前端所在的地址（a.com）。当我们请求后端文件时（也请求a.com）经过代理就会将实际的请求转到后端文件实际的地址b.com
