# 项目初始化
## 项目目录的创建
### git仓库的建立
1. 先在github中建立一个repository要点选gitignore功能。
2. 配置生成ssh key
3. 在本地设置远端仓库的用户名及密码

> git config --global user.name “XXX"

> git config --global user.email “XXXX"

#### git项目建立
1. 在项目文件夹运行> git init 建立本地的git仓库
2. 在本地将远程仓库添加到本地仓库

> git remote add origin ......(项目ssh地址).....

3. 将远程的仓库pull到本地

> git pull origin master

4. 可以使用git status来查看仓库的状态
5. <font color=#FF0033>编辑项目</font>
6. 将编辑好的项目添加到本地缓存

> git add .

7. 将缓存提交并添加注释（提交注释）

> git commit -am 'test'

8. 提交到远端

> git push --set-upstream origin master

#### gitignore设计
<em>gitignore用来设置不上传的文件</em>
在工程目录下建立.gitignore文件，打开即可设置。
一般会存在的文件有：

```python
.DS_Store
/node_modules/
/dist/
```

### 目录结构的设计

按照分层的设计来搭建目录：
- 主文件目录在src目录下
- 逻辑层：pages+view(页面+脚本&样式）将模板与静态文件分离
- 数据层:service
- 工具层:utils
- 图片&静态资源层:images
